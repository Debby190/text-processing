```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.nike.com/id/w/womens-long-sleeve-shirts-4cvaez5e1x6'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
response = requests.get(url, headers=headers)


if response.status_code == 200:
    # Parsing HTML
    bs = BeautifulSoup(response.content, 'html.parser')

    products = bs.find_all('div', class_='product-card__body')

    titles = []
    prices = []
    links = []

    results = []
    for product in products:
        title = product.find('div', class_='product-card__title').text.strip() if product.find('div', class_='product-card__title') else 'No Title'
        price = product.find('div', class_='product-price').text.strip() if product.find('div', class_='product-price') else 'No Price'
        link_tag = product.find('a', class_='product-card__link-overlay')
        link = link_tag['href'] if link_tag else 'No Link'

        product_data = {
            'title': title,
            'price': price,
            'link': f"https://www.nike.com{link}"
        }
        results.append(product_data)

    for product in results:
        print(f"Nama Produk: {product['title']}")
        print(f"Harga: {product['price']}")
        print(f"Link: {product['link']}")
        print("---")
else:
    print(f"Gagal mengakses halaman. Status code: {response.status_code}")
```
